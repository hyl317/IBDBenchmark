```python
import h5py
import numpy as np
import itertools

for ch in np.arange(1,23):
    print(f'chromosome {ch}')
    path2HDF5 = f'/mnt/archgen/users/yilei/IBDsim/groundtruth_IBD_glimpse_transversion/MAC1/hdf5/ch{ch}.h5'
    f = h5py.File(path2HDF5)
    for id1, id2 in itertools.combinations(np.arange(4), 2):
        gt1 = np.sum(f['calldata/GT'][:, id1, :], axis=1)
        gt2 = np.sum(f['calldata/GT'][:, id2, :], axis=1)
        homo_index = np.where(np.logical_and(np.logical_or(gt1==0, gt1==2), np.logical_or(gt2==0, gt2==2)))[0]
        oppohomo_index = np.where(np.logical_or(np.logical_and(gt1==0, gt2==2), np.logical_and(gt1==2, gt2==0)))[0]
        print(len(oppohomo_index)/len(homo_index))
```
